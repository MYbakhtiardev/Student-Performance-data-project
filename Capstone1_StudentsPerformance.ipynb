**<h2>Data Visulization</h2>**

**<h3>Distribution Plot</h3>**

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

subjects = ['math score', 'reading score', 'writing score']
colors = ['skyblue', 'lightcoral', 'lightgreen']

for i, subject in enumerate(subjects):
    sns.histplot(df[subject], bins=20, kde=True, ax=axes[i], color=colors[i])
    axes[i].set_title(f"Distribution of {subject}")
    axes[i].set_xlabel(subject.capitalize())
    axes[i].set_ylabel("Frequency")

    mean = df[subject].mean()
    median = df[subject].median()
    mode = df[subject].mode().values[0]

    axes[i].axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean:.2f}')
    axes[i].axvline(median, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median:.2f}')
    axes[i].axvline(mode, color='purple', linestyle='dashed', linewidth=1, label=f'Mode: {mode:.2f}')

    for line in axes[i].lines:
        line.set_linewidth(2)
        line.set_color('navy')

    axes[i].legend()

plt.tight_layout()
plt.show()
```

Note:

1.   Bars (Histogram) representing the frequency of scores in different ranges.
2.   Smooth Curve (Kernel Density Estimation - KDE) showing the probability distribution.
3.   Vertical Lines indicating:


*   Mean (blue dashed line, labeled) → The average score.
*   Median (another blue dashed line, labeled) → The middle value when sorted.
*   Mode (third blue dashed line, labeled) → The most frequently occurring score.











**<h3>Boxplot (Before Remove the Outliers)</h3>**

```python
subject_types = ['math score', 'reading score', 'writing score']
df_subject = df[subject_types]

plt.figure(figsize=(14, 9))
sns.boxplot(data=df_subject)
plt.title("Boxplot of Students Performance (Before Outlier Removal)")
plt.show()
```

Note:



1.  Box (Interquartile Range - IQR):
*   Represents the middle 50% of the data (from Q1 to Q3).
*   The horizontal line inside the box is the median (middle value).

2.  Whiskers:
*   Extend to the smallest and largest values within 1.5 times the IQR.

3.  Outliers (Circles below the whiskers):
*   These are values that fall significantly outside the normal range.
*   Observed in Math, Reading, and Writing scores, indicating some students scored very low.













**<h3>Remove the Outliers</h3>**

```python
subject_columns = ['math score', 'reading score', 'writing score']
df_subject = df[subject_columns]

Q1 = df_subject.quantile(0.25)
Q3 = df_subject.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_subject_filtered = df_subject[~((df_subject < lower_bound) | (df_subject > upper_bound)).any(axis=1)]
```

Note:



1.   Removes rows that contain any outlier in math, reading, or writing scores.
2.   Keeps only students with scores within the expected range.
3.   Creates a new filtered dataframe (df_subject_filtered) without outliers.


**<h3>Boxplot (After Remove the Outliers)</h3>**

```python
plt.figure(figsize=(14, 9))
sns.boxplot(data=df_subject_filtered)
plt.title("Boxplot of Blood Donations (After Outlier Removal)")
plt.show()
```

Note:

Note:



1.  Box (Interquartile Range - IQR):
*   The box represents the middle 50% of the data (from Q1 to Q3).
*   The line inside the box is the median (middle value of the dataset).

2.  Whiskers:
*   The whiskers extend to the lowest and highest values within the acceptable range (1.5 * IQR).
*   This helps to show the spread of scores without extreme outliers.

3.  Outliers Removed:
*   Compared to the previous boxplot (before outlier removal), the extreme low and high values (outliers) have been removed.
*   This ensures the analysis focuses on the majority of students' scores without being influenced by rare extreme values.





```python
df_final = df.loc[df_subject_filtered.index]
df_final
# After remove the outliers, from 1000 rows decrease to 988 rows
```

*   A strong positive correlation: Students who perform well in math tend to score high in reading.

3.   Reading vs. Writing (0.95)

*   A very strong correlation: Students with high reading scores generally excel in writing.
*   This suggests that reading and writing skills are closely related.

4.   Math vs. Writing (0.78)

*   A moderately strong correlation, but lower than reading-writing.
*   Math is more distinct from language-based subjects but still shows a relationship.
